### Deploy all manner of skills and techniques including graphs and plots to obtain meaningful insight from the dataset

```python
plt.hist(EbenData['sulphates'], bins=20)
plt.xlabel('sulphates content')
plt.ylabel('Frequency')
plt.title('Distribution of Sulphates in Red Wine')
plt.show()
```

```python
plt.hist(EbenData['alcohol'], bins=20)
plt.xlabel('alcohol content')
plt.ylabel('Frequency')
plt.title('Distribution of alcohol content in Red Wine')
plt.show()
```

```python
plt.hist(EbenData['pH'], bins=20)
plt.xlabel('pH content')
plt.ylabel('Frequency')
plt.title('Distribution of pH content in Red Wine')
plt.show()
```

##### Using Scatter Plots To Analyse The data

```python
plt.scatter(EbenData['alcohol'], EbenData['pH'])
plt.xlabel('Alcohol Content')
plt.ylabel('pH Level')
plt.title('Alcohol Content vs. pH Level')
plt.show()
```

```python
plt.scatter(EbenData['chlorides'], EbenData['density'])
plt.xlabel('chlorides')
plt.ylabel('density')
plt.title('chlorides vs. density')
plt.show()
```
